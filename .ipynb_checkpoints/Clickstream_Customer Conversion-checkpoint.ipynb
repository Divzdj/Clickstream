```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.feature_selection import VarianceThreshold
import warnings
warnings.filterwarnings("ignore")
```

## 1. Data Preprocessing

```python
#Load Datasets
train=pd.read_csv("data/train_data.csv")
test=pd.read_csv("data/test_data.csv")
print("Train Shape:",train.shape)
print("Test Shape:",test.shape)
```

```python
# Handling Missing Values
# Separate numerical and categorical columns
num_cols=train.select_dtypes(include=["int64","float64"]).columns
cat_cols=train.select_dtypes(include=["object"]).columns
# Fill missing values in train
for col in num_cols:
    train[col] = train[col].fillna(train[col].median())
for col in cat_cols:
    train[col] = train[col].fillna(train[col].mode()[0])
# Fill missing values in test
for col in num_cols:
    test[col] = test[col].fillna(test[col].median())
for col in cat_cols:
    test[col] = test[col].fillna(test[col].mode()[0])
```

```python
# Time-based Analysis for both train and test 
def add_time_features(df):
    # Create datetime column
    df["date"] = pd.to_datetime(df[["year", "month", "day"]])
    
    # Extract day of week
    df["day_of_week"] = df["date"].dt.day_name()
    
    # Browsing Duration & Page Views per session
    session_time = df.groupby("session_id").agg(
        browsing_duration=("order", lambda x: x.max() - x.min() + 1),
        page_views=("order", "count")
    ).reset_index()
    
    # Merge back
    df = df.merge(session_time[["session_id", "browsing_duration"]], 
                  on="session_id", how="left")
    
    return df

# Apply to both datasets
train = add_time_features(train)
test = add_time_features(test)

# Preview
print(train[["date", "day_of_week", "browsing_duration"]].head())
print(test[["date", "day_of_week", "browsing_duration"]].head())
```

## 3. Feature Engineering

```python
# Feature Engineering Function
def feature_engineering(df):
    #  SESSION METRICS 
    session_metrics = df.groupby("session_id").agg(
        session_length=("order", lambda x: x.max() - x.min() + 1),  # clicks span
        num_clicks=("order", "count"),  # total clicks
    ).reset_index()

    # Time spent per product category 
    time_per_category = df.groupby(["session_id", "page1_main_category"]).agg(
        clicks_in_category=("order", "count")
    ).reset_index()

    # Pivot so each category becomes a column
    time_per_category = time_per_category.pivot(
        index="session_id", columns="page1_main_category", values="clicks_in_category"
    ).fillna(0).reset_index()

    # Merge session metrics + category time
    df = df.merge(session_metrics, on="session_id", how="left")
    df = df.merge(time_per_category, on="session_id", how="left")

    # CLICKSTREAM PATTERNS 
    click_sequences = df.groupby("session_id")["page"].apply(list).reset_index()
    click_sequences.rename(columns={"page": "click_sequence"}, inplace=True)

    df = df.merge(click_sequences, on="session_id", how="left")

    #  BEHAVIORAL METRICS 
    # Bounce rate flag (1 if only 1 page view in session)
    df["is_bounce"] = df["num_clicks"] == 1

    # Exit page (last page in session)
    exit_pages = df.groupby("session_id")["page"].last().reset_index()
    exit_pages.rename(columns={"page": "exit_page"}, inplace=True)
    df = df.merge(exit_pages, on="session_id", how="left")

    # Revisit pattern: count how many times a session_id appears in dataset
    revisit_counts = df.groupby("session_id").size().reset_index(name="revisit_count")
    df = df.merge(revisit_counts, on="session_id", how="left")

    return df

# Apply to both train and test
train = feature_engineering(train)
test = feature_engineering(test)

# Quick check
print(train[["session_id", "session_length", "num_clicks", "is_bounce", "exit_page"]].head())
```

## 4. Final Preprocessing for Modelling

```python
#  Drop helper/EDA-only columns
for df in (train, test):
    for col in ['price_bin','date']:  
        if col in df.columns:
            df.drop(columns=[col], inplace=True)

#  Identify categorical columns present in BOTH train and test
cat_cols = [
    col for col in train.select_dtypes(include=['object', 'category']).columns
    if col in test.columns
]

#  Convert unhashable types (like lists) to strings 
for col in cat_cols:
    train[col] = train[col].apply(lambda x: str(x) if isinstance(x, list) else x)
    test[col] = test[col].apply(lambda x: str(x) if isinstance(x, list) else x)

#  Label encode binary categoricals 
le = LabelEncoder()
for col in cat_cols:
    if train[col].nunique() <= 2:
        le.fit(train[col].astype(str))
        train[col] = le.transform(train[col].astype(str))
        test[col] = test[col].astype(str).map(
            lambda x: le.transform([x])[0] if x in le.classes_ else 0
        )

#  One-hot encode remaining categoricals 
remaining_cat_cols = [col for col in cat_cols if train[col].nunique() > 2]
train = pd.get_dummies(train, columns=remaining_cat_cols, drop_first=True)
test = pd.get_dummies(test, columns=remaining_cat_cols, drop_first=True)

#  Align columns between train and test ---
train, test = train.align(test, join="left", fill_value=0)

#  Save encoded base datasets for all tasks 
train_encoded = train.copy()
test_encoded = test.copy()

def downcast_df(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='float')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='integer')
    return df

train_encoded = downcast_df(train_encoded)
test_encoded = downcast_df(test_encoded)
```

```python
#  Create task-specific splits 
## Classification
X_train_class = train_encoded.drop('price_2', axis=1)
y_train_class = train_encoded['price_2'].astype(int)

X_test_class = test_encoded.drop('price_2', axis=1)
y_test_class = test_encoded['price_2'].astype(int)
```

```python
## Regression 
X_train_reg = train_encoded.drop('price', axis=1)
y_train_reg = train_encoded['price']
X_test_reg = test_encoded.drop('price', axis=1)
y_test_reg = test_encoded['price']
```

```python
## Clustering (no target)
X_clu = train_encoded.drop(['price_2', 'price'], axis=1, errors='ignore')
```

## 5.Balancing Techniques (For Classification Models)

```python

#  Align columns between train and test 
X_train_class, X_test_class = X_train_class.align(X_test_class, join='left', axis=1, fill_value=0)

#  Ensure numeric features and string column names 
X_train_class = X_train_class.apply(pd.to_numeric, errors='coerce').fillna(0)
X_test_class = X_test_class.apply(pd.to_numeric, errors='coerce').fillna(0)
X_train_class.columns = X_train_class.columns.astype(str)
X_test_class.columns = X_test_class.columns.astype(str)

#  Optional: Drop zero-variance features to speed up SMOTE 
selector = VarianceThreshold(threshold=0.0)
X_train_reduced = selector.fit_transform(X_train_class)

#  Apply SMOTE with reduced neighbors for speed 
smote = SMOTE(random_state=42, n_jobs=-1, k_neighbors=3)
X_train_bal, y_train_bal = smote.fit_resample(X_train_reduced, y_train_class)

#  Encode labels consistently 
le = LabelEncoder()
le.fit(pd.concat([pd.Series(y_train_bal), pd.Series(y_test_class)], axis=0))
y_train_bal = le.transform(y_train_bal)
y_test_class = le.transform(y_test_class)


```

```python
#  Align columns between train and test 
X_train_class, X_test_class = X_train_class.align(X_test_class, join='left', axis=1, fill_value=0)

#  Ensure numeric features and string column names 
X_train_class = X_train_class.apply(pd.to_numeric, errors='coerce').fillna(0)
X_test_class = X_test_class.apply(pd.to_numeric, errors='coerce').fillna(0)
X_train_class.columns = X_train_class.columns.astype(str)
X_test_class.columns = X_test_class.columns.astype(str)

#  Drop zero-variance features to speed up SMOTE 
selector = VarianceThreshold(threshold=0.0)
X_train_reduced = selector.fit_transform(X_train_class)
selected_columns = X_train_class.columns[selector.get_support()]  # Only retained columns

#  Apply SMOTE with reduced neighbors 
smote = SMOTE(random_state=42, n_jobs=-1, k_neighbors=3)
X_train_bal, y_train_bal = smote.fit_resample(X_train_reduced, y_train_class)

#  Encode labels consistently 
le = LabelEncoder()
le.fit(pd.concat([pd.Series(y_train_bal), pd.Series(y_test_class)], axis=0))
y_train_bal = le.transform(y_train_bal)
y_test_class = le.transform(y_test_class)
```

```python
def scale_with_standardscaler(X_train, X_test, drop_low_variance=True, variance_threshold=1e-6, sample_size=None):
    # --- Optional sampling for speed ---
    if sample_size:
        X_train = X_train.sample(n=sample_size, random_state=42)
        X_test = X_test.sample(n=min(sample_size, len(X_test)), random_state=42)

    # --- Ensure inputs are DataFrames ---
    if not isinstance(X_train, pd.DataFrame) or not isinstance(X_test, pd.DataFrame):
        raise ValueError("Inputs must be pandas DataFrames.")

    # --- Downcast numeric types ---
    for df in [X_train, X_test]:
        df[df.select_dtypes('float64').columns] = df.select_dtypes('float64').apply(pd.to_numeric, downcast='float')
        df[df.select_dtypes('int64').columns] = df.select_dtypes('int64').apply(pd.to_numeric, downcast='integer')

    # --- Detect numeric columns ---
    num_cols = X_train.select_dtypes(include='number').columns

    # --- Drop low-variance columns (vectorized) ---
    if drop_low_variance:
        stds = X_train[num_cols].std(axis=0)
        high_var_cols = stds[stds > variance_threshold].index.tolist()
    else:
        high_var_cols = num_cols.tolist()

    # --- Scale only high-variance columns ---
    scaler = StandardScaler()
    X_train_scaled = X_train.copy()
    X_test_scaled = X_test.copy()
    X_train_scaled[high_var_cols] = scaler.fit_transform(X_train[high_var_cols])
    X_test_scaled[high_var_cols] = scaler.transform(X_test[high_var_cols])

    return X_train_scaled, X_test_scaled, scaler
```

```python
# Classification branch
# After SMOTE
# --- Convert SMOTE output to DataFrame using selected columns ---
X_train_bal_df = pd.DataFrame(X_train_bal, columns=selected_columns)

# --- Match test set to reduced train columns ---
X_test_selected = X_test_class[selected_columns]

# --- Scale both sets using StandardScaler ---
X_train_class_scaled, X_test_class_scaled, scaler_class = scale_with_standardscaler(
    X_train_bal_df,
    X_test_selected
)
```

```python
# Regression branch 
X_train_reg = train_encoded.drop('price', axis=1)
y_train_reg = train_encoded['price']

X_test_reg = test_encoded.drop('price', axis=1)
y_test_reg = test_encoded['price']

X_train_reg_scaled, X_test_reg_scaled, scaler_reg = scale_with_standardscaler(
    X_train_reg,
    X_test_reg
)
```

```python
#  Clustering branch
X_clu = train_encoded.drop(['price_2', 'price'], axis=1, errors='ignore')

X_clu_scaled, _, scaler_clu = scale_with_standardscaler(
    X_clu,
    X_clu  # no separate test set for clustering
)
```

## 5. Model Building:


```python
#Supervised Learning Models
#Classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# --- Step 1: Remap training labels so they start at 0 ---
unique_classes = np.unique(y_train_bal)
label_mapping = {old_label: new_label for new_label, old_label in enumerate(unique_classes)}
y_train_bal = np.array([label_mapping[label] for label in y_train_bal])

print("Remapped training labels:", np.unique(y_train_bal))

# --- Step 2: Convert test data to pandas objects for safe masking ---
y_test_class = pd.Series(y_test_class, name="target")
X_test_class_scaled = pd.DataFrame(X_test_class_scaled)

# --- Step 3: Create filtered test set without class 0 (for XGBoost only) ---
mask_no_zero = y_test_class != 0
X_test_class_no_zero = X_test_class_scaled.loc[mask_no_zero]
y_test_class_no_zero = y_test_class.loc[mask_no_zero]

print("Unique labels in filtered test set:", np.unique(y_test_class_no_zero))
print("Filtered test set shapes:", X_test_class_no_zero.shape, y_test_class_no_zero.shape)
```

```python
# --- Define models ---
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "XGBoost": XGBClassifier(
        n_estimators=200,
        learning_rate=0.1,
        max_depth=6,
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss'
    ),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=300, random_state=42)
}

# --- Train, predict, and evaluate ---
results = {}

for name, model in models.items():
    print(f"\n=== {name} ===")
    
    # Special handling for XGBoost to avoid unseen class 0
    if name == "XGBoost":
        model.fit(X_train_class_scaled, y_train_bal)
        y_pred = model.predict(X_test_class_no_zero)
        acc = accuracy_score(y_test_class_no_zero, y_pred)
        print(f"Accuracy (classes 1 & 2 only): {acc:.4f}")
        print(classification_report(y_test_class_no_zero, y_pred))
    else:
        model.fit(X_train_class_scaled, y_train_bal)
        y_pred = model.predict(X_test_class_scaled)
        acc = accuracy_score(y_test_class, y_pred)
        print(f"Accuracy: {acc:.4f}")
        print(classification_report(y_test_class, y_pred))
    
    results[name] = acc

# --- Summary table ---
results_df = pd.DataFrame(list(results.items()), columns=['Model', 'Accuracy'])
results_df = results_df.sort_values(by='Accuracy', ascending=False)
print("\nModel Performance Summary:")
print(results_df)
```
