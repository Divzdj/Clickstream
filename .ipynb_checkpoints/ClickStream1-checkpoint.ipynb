# **1. Data Cleaning**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
import cloudpickle
```

```python
#Load Datasets
train=pd.read_csv("data/train_data.csv")
test=pd.read_csv("data/test_data.csv")
print("Train Shape:",train.shape)
print("Test Shape:",test.shape)
```

```python
 #Fill numeric missing values with median
for col in train.select_dtypes(include=['float64','int64']).columns:
    train[col].fillna(train[col].median(), inplace=True)
    test[col].fillna(test[col].median(), inplace=True)

# Fill categorical missing values with mode
for col in train.select_dtypes(include=['object']).columns:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(test[col].mode()[0], inplace=True)

print(f"Number of duplicate rows: {train.duplicated().sum()}")
```

# **2.Feature Engineering**

```python
for df in [train, test]:
    # Session-level metrics
    df['session_length'] = df.groupby('session_id')['order'].transform('max')
    df['num_clicks'] = df.groupby('session_id')['order'].transform('count')
    df['time_per_category'] = df.groupby(['session_id','page1_main_category'])['order'].transform('count')
    df['first_page'] = df.groupby('session_id')['page'].transform('first')
    df['last_page'] = df.groupby('session_id')['page'].transform('last')
    df['unique_categories'] = df.groupby('session_id')['page1_main_category'].transform('nunique')
    df['is_bounce'] = np.where(df['session_length']==1,1,0)
    df['exit_page'] = df['last_page']
    df['is_revisit'] = df.groupby(['session_id','page2_clothing_model'])['page'].transform(lambda x: 1 if x.duplicated().any() else 0)
```

```python
# Summarize per session
session_summary = train.groupby('session_id').agg({
    'session_length':'max',
    'num_clicks':'max',
    'time_per_category':'sum',
    'unique_categories':'max',
    'is_bounce':'mean',
    'is_revisit':'max'
}).reset_index()

print(session_summary.head())
```

```python
# Copy dataset for EDA
eda_df = train.copy()
```

```python
# Check categorical columns
categorical_cols = train.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    if col in test.columns:
        mapping = {cls: idx for idx, cls in enumerate(le.classes_)}
        test[col] = test[col].map(mapping)
        test[col].fillna(-1, inplace=True)
        test[col] = test[col].astype(int)
```

```python
# Non-feature columns
non_feature_cols = ['session_id','price','price_2']

# Classification features
X_class = train.drop(columns=non_feature_cols)
y_class = train['price_2'].astype(int)

# Regression features
X_reg = train.drop(columns=['session_id','price'], errors='ignore')
y_reg = train['price']
```

```python
# Clustering features
X_clu = train.drop(columns=['session_id','price','price_2'], errors='ignore')
```

```python
print(X_class.columns[:20])  # should NOT include price, price_2, session_id
```

```python
# --- Save feature lists for app.py ---
pd.Series(X_class.columns).to_csv("class_features.csv", index=False)
pd.Series(X_reg.columns).to_csv("reg_features.csv", index=False)
pd.Series(X_clu.columns).to_csv("clu_features.csv", index=False)
print("Feature lists saved: class_features.csv, reg_features.csv, clu_features.csv")
```

# **4. Classification Pipeline with Smote**


```python
#  Map target to 0 and 1 (XGBoost requires labels starting from 0)
y_class_mapped = y_class.map({1: 0, 2: 1})

# Display target counts BEFORE SMOTE
print("Target distribution BEFORE SMOTE:")
print(Counter(y_class_mapped))

plt.figure(figsize=(5,4))
sns.countplot(x=y_class_mapped.astype(str))
plt.title("PRICE_2 Distribution Before SMOTE")
plt.show()


# Split into train and validation sets
Xc_train, Xc_val, yc_train, yc_val = train_test_split(
    X_class, y_class_mapped, test_size=0.2, random_state=42
)


# Define models
models = {
    'LogisticRegression': LogisticRegression(max_iter=500, class_weight='balanced', random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'RandomForest': RandomForestClassifier(random_state=42, class_weight='balanced'),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'NeuralNetwork': MLPClassifier(max_iter=500, random_state=42)
}

# Hyperparameter grids for GridSearch
params = {
    'DecisionTree': {'model__max_depth':[5,10,None],'model__min_samples_split':[2,5]},
    'RandomForest': {'model__n_estimators':[100,200],'model__max_depth':[5,10,None]},
    'XGBoost': {'model__n_estimators':[100,200],'model__max_depth':[3,5],'model__learning_rate':[0.01,0.1]},
    'NeuralNetwork': {'model__hidden_layer_sizes':[(50,50),(100,)], 'model__activation':['relu','tanh']}
}

results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    pipeline = ImbPipeline([
        ('smote', SMOTE(random_state=42)),
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    
    if name in params:
        grid = GridSearchCV(pipeline, params[name], cv=3, scoring='accuracy', n_jobs=-1)
        grid.fit(Xc_train, yc_train)
        best_model = grid.best_estimator_
        print(f"Best params for {name}: {grid.best_params_}")
    else:
        pipeline.fit(Xc_train, yc_train)
        best_model = pipeline

    # Validation predictions
    y_pred = best_model.predict(Xc_val)
    
    results[name] = {
        'Accuracy': accuracy_score(yc_val, y_pred),
        'Precision': precision_score(yc_val, y_pred, average='weighted', zero_division=0),
        'Recall': recall_score(yc_val, y_pred, average='weighted'),
        'F1-Score': f1_score(yc_val, y_pred, average='weighted')
    }

# Results
results_df = pd.DataFrame(results).T
print("\nClassification Results:\n", results_df)

# Save best model
with open("best_classifier.pkl","wb") as f:
    cloudpickle.dump(best_model, f)

# Display target counts AFTER SMOTE
smote = SMOTE(random_state=42)
X_class_res, y_class_res = smote.fit_resample(X_class, y_class_mapped)

print("\nTarget distribution AFTER SMOTE:")
print(Counter(y_class_res))

plt.figure(figsize=(5,4))
sns.countplot(x=y_class_res.astype(str))
plt.title("PRICE_2 Distribution After SMOTE")
plt.show()
```

# **5. Regression Pipeline**

```python
regressors = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0, random_state=42),
    "Lasso": Lasso(alpha=0.01, random_state=42),
    "GradientBoosting": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
regression_results = {}

for name, model in regressors.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipeline.fit(X_reg_train, y_reg_train)
    y_pred = pipeline.predict(X_reg_test)
    
    regression_results[name] = {
        "MAE": mean_absolute_error(y_reg_test, y_pred),
        "MSE": mean_squared_error(y_reg_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_reg_test, y_pred)),
        "R2": r2_score(y_reg_test, y_pred)
    }

pd.DataFrame(regression_results).T

best_regressor = Pipeline([('scaler', StandardScaler()), ('model', GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42))])
best_regressor.fit(X_reg, y_reg)

with open("best_regressor.pkl","wb") as f:
    cloudpickle.dump(best_regressor, f)
```

# **6.Clustering Pipeline**

```python
# Sample data to reduce memory
X_clu_sample = X_clu.sample(n=min(10000, len(X_clu)), random_state=42)

# --- KMeans ---
wcss = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_clu_sample)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(2,11), wcss, marker='o')
plt.title("Elbow Method for KMeans")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X_clu_sample)
print("KMeans Silhouette Score:", silhouette_score(X_clu_sample, kmeans_labels))
print("KMeans Davies-Bouldin Index:", davies_bouldin_score(X_clu_sample, kmeans_labels))
```

```python
# --- DBSCAN ---
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_clu_sample)
mask = dbscan_labels != -1
if np.sum(mask) > 1:
    print("DBSCAN Silhouette Score:", silhouette_score(X_clu_sample[mask], dbscan_labels[mask]))
    print("DBSCAN Davies-Bouldin Index:", davies_bouldin_score(X_clu_sample[mask], dbscan_labels[mask]))
else:
    print("DBSCAN: Not enough clusters for silhouette score")
```

```python
# --- Agglomerative + PCA pipeline ---
pca = PCA(n_components=10, random_state=42)
X_clu_reduced = pca.fit_transform(X_clu_sample)
agglo = AgglomerativeClustering(n_clusters=7)
agglo_labels = agglo.fit_predict(X_clu_reduced)
print("Agglomerative Silhouette Score:", silhouette_score(X_clu_reduced, agglo_labels))
print("Agglomerative Davies-Bouldin Index:", davies_bouldin_score(X_clu_reduced, agglo_labels))
```

```python
with open("dbscan_model.pkl", "wb") as f:
    cloudpickle.dump(dbscan, f)
```
